```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the merged dataset
file_path = "merged_sales_data.csv"  # Update with the correct path
df = pd.read_csv(file_path)

# Convert 'datum' column to datetime format
df["datum"] = pd.to_datetime(df["datum"])

# Display dataset info
print(df.info())
```

```python
# Identify numerical columns
num_cols = df.select_dtypes(include=["float64", "int64"]).columns

# Compute Q1, Q3, and IQR
Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find number of outliers per column
outliers = ((df[num_cols] < lower_bound) | (df[num_cols] > upper_bound)).sum()
print("Outliers per column:\n", outliers)
# Apply clipping for each column separately
for col in num_cols:
    df[col] = df[col].clip(lower=lower_bound[col], upper=upper_bound[col])
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[num_cols])
plt.xticks(rotation=90)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# Filter only numerical columns except 'Year', 'Month', 'Hour'
num_cols_filtered = [col for col in num_cols if col not in ["Year", "Month", "Hour"]]

# Apply Standardization (Z-score normalization)
scaler = StandardScaler()
df[num_cols_filtered] = scaler.fit_transform(df[num_cols_filtered])
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[num_cols_filtered])
plt.xticks(rotation=90)
plt.show()
```
